```python
import pandas as pd
from datetime  import datetime
data = pd.read_csv("voitures.csv", sep=";")
data.columns=["immat","marque","kilom","date circulation"]
data
x_=data.copy()
x_.set_index("immat", inplace=True)
x_
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder()
ohe.fit(x_[['marque']])
ohe.categories_[0]
marque_encoded=ohe.transform(x_[['marque']])
marque_encoded=marque_encoded.toarray()
marque_encoded
df_marque_encoded=pd.DataFrame(marque_encoded,
                                    index=x_.index,
                                    columns=ohe.categories_[0])
df_marque_encoded
X_=pd.concat([x_,df_marque_encoded], axis=1)
X_
X_.drop('marque', axis=1, inplace=True, errors='ignore')

X_['date circulation']=pd.to_datetime(X_['date circulation'])
X_['date circulation']
X_['anciennete']=X_['date circulation'].apply(lambda date : datetime.now().year-date.year)
X_
X_.drop('date circulation', axis=1, inplace=True, errors='ignore')
X_
```
